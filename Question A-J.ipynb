```python
# Preparation
# Required packages/modules
import csv
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# a. First method- conversion of txt to csv w/ csv package
with open('horse-colic.txt', 'r') as in_file:
    stripped = (line.strip() for line in in_file)
    lines = (line.split(',') for line in stripped if line)
    with open('horse-colic-one.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerows(lines)
```

```python
# a. Second method- conversion of txt to csv w/ pandas package
df_rawTwo = pd.read_csv('horse-colic.txt', sep = ',', header = None)
df_rawTwo.to_csv('horse-colic-two.csv', index = None, header = None)
```

```python
# b. Demonstration of first method
df_rawOne = pd.read_csv('horse-colic-one.csv', header = None)
df_rawOne
```

```python
# b. Demonstration of second method
df_rawTwoTrial = pd.read_csv('horse-colic-two.csv', header = None)
df_rawTwoTrial
```

```python
# c. Replacement of ? values w/ NaN
# Presently, horse-colic-one is used
df_rawThree = df_rawOne.replace(to_replace = '?', value = 'NaN')
df_rawThree = df_rawThree.apply(pd.to_numeric, errors='coerce')
```

```python
# c. Demonstration of successful replacement of ? w/ NaN
df_rawThree
```

```python
# d. Saving latest dataset with NaN
df_rawThree.to_csv('horse-colic-with-NaN.csv', index = None, header = None)
```

```python
# e. Renaming header titles with 'horse-colic-with-NaN'
df_withNaN = pd.read_csv('horse-colic-with-NaN.csv', names=['surgery', 'age','hosp_no','rect_temp','pulse', 'resp_rate','ex_temp','pr_pulse','muc_mb','crt','pain','prss','abd_dist','ns_tube','ns_flux','ns_flux_PH','rect_exF','abdomen','pcv','tp','abd_app','abd_tp','outcome','les_surg','les_site','les_type','les_sType','les_code'])
df_withNaN
```

```python
# f. Saving latest dataset with header titles
df_withNaN.to_csv('horse-colic-with-title.csv', index = None)
```

```python
# g. Loading 'horse-colic-with-title' and display list of total NaN at each column
df_withTitle = pd.read_csv('horse-colic-with-title.csv')
count_nan = df_withTitle.isnull().sum()
print('Total NaN in Each Column:')
print(count_nan)
```

```python
# h. Visualisation of total NaN at each column
# Converting header titles to list for y-axis of horizontal bar chart
columnLabels = list(df_withTitle.columns)
# Horizontal bar chart preparation
plt.figure(figsize=(10, 10))
plt.barh(y = columnLabels, width = count_nan, height = 0.5, color = 'blue')
plt.xlabel('NaN Counts')
plt.ylabel('Columns')
plt.title('Counts of NaN in Horse Colic Dataset', loc = 'left')
plt.rcParams.update({'font.size': 10})
plt.grid(True)
plt.show()
```
